#### What product sold the most? Why do you think it sold the most?

```python
grouped_product = data_new.groupby('Product')
display(grouped_product.sum().sort_values(by='Quantity Ordered', ascending=False))

```

```python
#plot 
quantity_ordered_product = grouped_product.sum()['Quantity Ordered']
quantity_ordered_product

qua = [q for q, df in data_new.groupby('Product')]

plt.plot(qua, quantity_ordered_product)
plt.bar(qua, quantity_ordered_product)
# plt.plot(Hours, results3['Count'])
plt.xticks(qua, rotation='vertical')
#plt.grid()
plt.xlabel('Products')
plt.ylabel('Quantity Ordered')
plt.show()

### AAA Batteries (4-pack) comes first, followed by AA Batteries (4-pack)
```

```python
# We need also to see proce for each item to check is this the price is a reason for products sold the most
Price_grp = data_new.groupby('Product').mean().sort_values(by='Price Each', ascending=True)
Price_grp.head()

# It is clear that the products that sold most are the same products that have lowest price, so the price is one of the key reason
```

```python
# Referenced: https://stackoverflow.com/questions/14762181/adding-a-y-axis-label-to-secondary-y-axis-in-matplotlib

prices = data_new.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(qua, quantity_ordered_product, color='g')
ax2.plot(qua, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(qua, rotation='vertical', size=8)

fig.show()
```
